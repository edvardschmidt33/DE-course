```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Create random data
df = pd.DataFrame(
np.random.rand(10, 4),
columns=['a'
,
'b'
,
'c'
,
'd']
)
# Plot as area chart
df.plot.area()
plt.title("Random Data Visualization")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```
